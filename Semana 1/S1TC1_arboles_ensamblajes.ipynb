### Punto 3 - Regresión lineal
En la celda 3 ajuste un modelo de regresión lineal a todo el conjunto de datos, utilizando "total" como variable de respuesta y "season" y "hour" como las únicas variables predictoras, teniendo en cuenta que la variable "season" es categórica. Luego, imprima los coeficientes e interprételos. ¿Cuáles son las limitaciones de la regresión lineal en este caso?

```python
# Celda 3.A
## Crea un dataframe para crear el modelo
df_reg = bikes[['season','hour','total']]
df_reg = df_reg.reset_index()
df_reg.drop(columns=['datetime'],inplace=True)
df_reg = pd.get_dummies(df_reg, columns=['season']).astype(int)

## Crea las columnas X y Y (Entrada y Salida respectivamente)
X_train_reg = df_reg[['hour','season_1','season_2','season_3','season_4']]
Y_traing_reg = df_reg[['total']]

## Entrena el modelo
modelo_reg = LinearRegression()
modelo_reg.fit(X_train_reg, Y_traing_reg)

## Mostrar los coeficientes del modelo ajustado
print("Coeficiente:", modelo_reg.coef_)
print("Intercepto:", modelo_reg.intercept_)

## Error de prediccion en entrenamiento
y_pred_train_reg = modelo_reg.predict(X_train_reg)
mse_train = mean_squared_error(Y_traing_reg, y_pred_train_reg)
r2_train = r2_score(Y_traing_reg, y_pred_train_reg)

print("Error cuadrático medio (MSE) en entrenamiento:", mse_train)
print("Coeficiente de determinación (R^2) en entrenamiento:", r2_train)
```

$Y_{\text{Total}} = -8.22579109e^{12} + 10.5451458*X_{\text{Hora}} + 8.22579109e^{12}*X_{\text{Invierno}} + 8.22579109e^{12}*X_{\text{Primavera}} + 8.22579109e^{12}*X_{\text{Verano}} + 8.22579109e^{12}*X_{\text{Otoño}}$

_Analisis:_

### Punto 4 - Árbol de decisión manual
En la celda 4 cree un árbol de decisiones para pronosticar la variable "total" iterando **manualmente** sobre las variables "hour" y  "season". El árbol debe tener al menos 6 nodos finales.

```python
# Celda 4
def calc_ecm(data,prom):
    N = len(data)
    ECM = ((data-prom)**2).sum()/N
    return ECM

def calc_div_cand(datos_array):
    ## funcion que permite calcular los divisiones candidatas
    array_sort = datos_array.sort_values()
    array_sort = array_sort.drop_duplicates()
    prom_consec = (array_sort + array_sort.shift(-1)) / 2
    prom_consec = prom_consec.dropna()
    return prom_consec.values

def calc_prom_each_part(X_data, Y_data, div_candidats):
    total_data_aux = X_data
    div_col = 0
    div_val = -1
    ecm_total_aux = -1
    part = 1
    for data_aux in total_data_aux:
        total_data = data_aux
        for n_col in range(len(div_candidats)):
            for split in range(len(div_candidats[n_col])):
                data_left = total_data[total_data[total_data.columns[n_col]] <= div_candidats[n_col][split]]
                data_right = total_data[total_data[total_data.columns[n_col]] > div_candidats[n_col][split]]
                prom_left = data_left.Target.mean()
                prom_right = data_right.Target.mean()
                ecm_left = calc_ecm(data_left[data_left.columns[n_col]],prom_left)
                ecm_right = calc_ecm(data_right[data_right.columns[n_col]],prom_right)
                ecm_total = (ecm_left*(len(data_left))/(len(data_left)+len(data_right))) + (ecm_right*(len(data_right))/(len(data_left)+len(data_right)))
                if div_val == -1:
                    div_col = n_col
                    div_val = div_candidats[n_col][split]
                    ecm_total_aux = ecm_total
                    prom_left_f = prom_left
                    prom_right_f = prom_right
                elif ecm_total_aux > ecm_total:
                    div_col = n_col
                    div_val = div_candidats[n_col][split]
                    ecm_total_aux = ecm_total
                    part_f = part
                    prom_left_f = prom_left
                    prom_right_f = prom_right
        part = part + 1
    return div_col,div_val,part_f,prom_left_f,prom_right_f

def arb_dec(X_data, Y_data, max_depth):
    X_data_array = []
    div_candidats = []
    ramificacion = []
    total_data = X_data.copy()
    total_data['Target'] = Y_data.copy()
    X_data_array.append(total_data)
    X_data_tmp = []
    for col in range(X_data.shape[1]):
        div_candidats.append(calc_div_cand(X_data[X_data.columns[col]]))
    for i in range(max_depth):
        X_data_array_aux = []
        div_col,div_val,part,prom_left_f,prom_right_f = calc_prom_each_part(X_data_array, Y_data, div_candidats)
        ramificacion.append([div_col,div_val,part,prom_left_f,prom_right_f])
        print("Se realiza la particion por la columna:",X_data.columns[div_col])
        print("La particion en la columna se hara en el valor:",div_val)
        X_data_tmp = X_data_array[part-1].copy()
        X_data_array.append(X_data_tmp[X_data_tmp[X_data_tmp.columns[div_col]] <= div_val])
        X_data_array.append(X_data_tmp[X_data_tmp[X_data_tmp.columns[div_col]] > div_val])
        del X_data_array[part-1]
    return ramificacion

df_arb = bikes[['season','hour','total']]
df_arb = df_arb.reset_index()
df_arb.drop(columns=['datetime'],inplace=True)

X_data = df_arb[['season','hour']]
Y_data = df_arb[['total']]

arbol = arb_dec(X_data,Y_data,max_depth=3)
print(arbol)
```

### Punto 5 - Árbol de decisión con librería
En la celda 5 entrene un árbol de decisiones con la **librería sklearn**, usando las variables predictoras "season" y "hour" y calibre los parámetros que considere conveniente para obtener un mejor desempeño. Recuerde dividir los datos en conjuntos de entrenamiento y validación para esto. Comente el desempeño del modelo con alguna métrica de desempeño de modelos de regresión y compare desempeño con el modelo del punto 3.

```python
# Celda 5
## Crea un dataframe para crear el modelo
df_dts = bikes[['season','hour','total']]
df_dts = df_dts.reset_index()
df_dts.drop(columns=['datetime'],inplace=True)

## Crea las columnas X y Y (Entrada y Salida respectivamente)
X_dts = df_reg[['hour','season']]
Y_dts = df_reg[['total']]

# se parte los datos en entrenamiento y test
X_train_dts, X_test_dts, y_train_dts, y_test_dts = train_test_split(X_dts, Y_dts, test_size=0.3, random_state=0)

##### Generacion del modelo
# Lista de valores para calibrar el criterio de parada de máxima profundidad (max_depth)
max_depth_range = range(1, 21)
# Lista para guardar los valores del RMSE para cada valor de máxima profundidad (max_depth)
MSE_scores = []
# Loop para obtener el desempeño del modelo de acuerdo con la máxima profundidad
for depth in max_depth_range:
    ## Crecion del modelo
    modelo_dts = DecisionTreeRegressor(max_depth=depth, random_state=1)
    # Realizar validación cruzada con 5 particiones
    scores = cross_val_score(modelo_dts, X_train_dts, y_train_dts, cv=5, scoring='neg_mean_squared_error').mean()
    # Convertir puntuaciones a errores cuadráticos medios positivos
    MSE_scores.append(-1*scores)
# Gráfica max_depth versus MSE (error del modelo)
plt.plot(max_depth_range, MSE_scores)
plt.xlabel('max_depth')
plt.ylabel('MSE')
```

Aca podemos tomar el modelo que tenga el valor de MSE mas bajo.

```python
# max_depth=8 fue el mejor, se ajusta un árbol usando este valor 
modelo_dts = DecisionTreeRegressor(max_depth=8, random_state=1)
modelo_dts.fit(X_train_dts, y_train_dts)

## Usar el modelo en el conjunto de prueba
y_pred_dts = modelo_dts.predict(X_test_dts)

# Calcular el MSE para los datos test
mse_test = mean_squared_error(y_test_dts, y_pred_dts)

print("Error cuadrático medio en el conjunto de prueba:", mse_test)
```

_Analisis:_  
El arbol de decision usando crossvalidation para tratar de encontrar el mejor modelo fue una muy buena opcion.  
Si se compara los resultados de este modelo (MSE 13275.45) y el modelo del ejercicio 3 que es una regresion lineal (MSE 25480.69), se observa que el error disminuye casi a la mitad con este arbol de decision.

### Punto 6 - Árbol de decisión y regresión logística
En la celda 6 construya un árbol de decisión y una regresión logística. Para el árbol calibre al menos un parámetro y evalúe el desempeño de cada modelo usando las métricas de Accuracy y F1-Score.

#### Se inicia con el arbol de decisiones:

```python
# Celda 6
##### Generacion del modelo del Arbol
# Lista de valores para calibrar el criterio de parada de máxima profundidad (max_depth)
max_depth_range = range(1, 25)
# Lista para guardar los valores de precision para cada valor de máxima profundidad (max_depth)
accuracy_scores = []
f1_scores = []
# Loop para obtener el desempeño del modelo de acuerdo con la máxima profundidad
for depth in max_depth_range:
    ## Crecion del modelo
    modelo_dtsc = DecisionTreeClassifier(max_depth=depth, random_state=1)
    # Realizar validación cruzada con 5 particiones
    accuracy_scores.append(cross_val_score(modelo_dtsc, X_train, y_train, cv=10, scoring='accuracy').mean())
    f1_scores.append(cross_val_score(modelo_dtsc, X_train, y_train, cv=10, scoring='f1_weighted').mean())
```

```python
# Gráfica max_depth versus MSE (error del modelo)
plt.plot(max_depth_range, accuracy_scores, label='Accuracy')
plt.plot(max_depth_range, f1_scores, label='f1_score')
plt.xlabel('max_depth')
plt.ylabel('score')
plt.legend()
```

```python
# El modelo con mejor valor f1-score y accuracy es de profundidad 2
modelo_dtsc = DecisionTreeClassifier(max_depth=2, random_state=1)
modelo_dtsc.fit(X_train, y_train)
# Realizar predicciones sobre el conjunto de prueba
y_pred_dtsc = modelo_dtsc.predict(X_test)
# Calcular la precisión del modelo
accuracy_dtsc = accuracy_score(y_test, y_pred_dtsc)
# Calcula el f1-score
f1_dtsc = f1_score(y_test, y_pred_dtsc, average='weighted')
print("El modelo de regresion logistica tuvo un valor de accuracy de:",accuracy_dtsc)
print("El modelo de regresion logistica tuvo un valor de af1-score de:",f1_dtsc)
```

#### Ahora se trabaja sobre la regresion logistica:

```python
# Se construye un modelo de regresion logistica
modelo_reglog = LogisticRegression()
modelo_reglog.fit(X_train, y_train)
# Realizar predicciones sobre el conjunto de prueba
y_pred_reglog = modelo_reglog.predict(X_test)
# Calcular la precisión del modelo
accuracy_reglog = accuracy_score(y_test, y_pred_reglog)
# Calcula el f1-score
f1_reglog = f1_score(y_test, y_pred_reglog, average='weighted')
print("El modelo de regresion logistica tuvo un valor de accuracy de:",accuracy_reglog)
print("El modelo de regresion logistica tuvo un valor de f1-score de:",f1_reglog)
```

### Punto 7 - Votación Mayoritaria
En la celda 7 elabore un esamble con la metodología de **Votación mayoritaria** compuesto por 300 muestras bagged para cada uno de los siguientes escenarios:

-100 árboles de decisión donde max_depth tome un valor de su elección\
-100 árboles de decisión donde min_samples_leaf tome un valor de su elección\
-100 regresiones logísticas

Evalúe cada uno de los tres modelos de manera independiente utilizando las métricas de Accuracy y F1-Score, luego evalúe el ensamble de modelos y compare los resultados. 

```python
# Celda 7
# Definición del modelo de arbol de decision con max_depth definido
from sklearn.ensemble import BaggingClassifier
clf_1 = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=20), n_estimators=100, bootstrap=True,
                        random_state=42, n_jobs=-1, oob_score=True)
# Definición del modelo de arbol de decision con min_samples_leaf definido
from sklearn.ensemble import BaggingClassifier
clf_2 = BaggingClassifier(base_estimator=DecisionTreeClassifier(min_samples_leaf=33), n_estimators=100, bootstrap=True,
                        random_state=42, n_jobs=-1, oob_score=True)
# Definición del modelo de regresion logistica
from sklearn.ensemble import BaggingClassifier
clf_3 = BaggingClassifier(base_estimator=LogisticRegression(max_iter=100000), n_estimators=100, bootstrap=True,
                        random_state=42, n_jobs=-1, oob_score=True)
```

Se realiza la prediccion de los 3 modelos creados anteriormente con Bagging de manera independiente:

```python
# Predicción y desempeño del primer modelo
clf_1.fit(X_train, y_train)
y_pred_1 = clf_1.predict(X_test)
print('La metrica F1 para el primer modelo:',metrics.f1_score(y_pred_1, y_test))
print('La metrica Accuracy para el primer modelo:',metrics.accuracy_score(y_pred_1, y_test))

# Predicción y desempeño del segundo modelo
clf_2.fit(X_train, y_train)
y_pred_2 = clf_2.predict(X_test)
print('La metrica F1 para el primer modelo:',metrics.f1_score(y_pred_2, y_test))
print('La metrica Accuracy para el primer modelo:',metrics.accuracy_score(y_pred_2, y_test))

# Predicción y desempeño del tercer modelo
clf_3.fit(X_train, y_train)
y_pred_3 = clf_3.predict(X_test)
print('La metrica F1 para el primer modelo:',metrics.f1_score(y_pred_3, y_test))
print('La metrica Accuracy para el primer modelo:',metrics.accuracy_score(y_pred_3, y_test))
```

Ahora se realiza la prediccion con votacion de manera ensamblada:

```python
# Crear VotingClassifier combinando los modelos entrenados
clf_vot_May = VotingClassifier(estimators=[('clf_1', clf_1), ('clf_2', clf_2), ('clf_3', clf_3)], voting='hard')

# Entrenar VotingClassifier
clf_vot_May.fit(X_train, y_train)

# Hacer predicciones
y_pred_vot_May = clf_vot_May.predict(X_test)

# Se calcula la metrica F1 y Accuracy
print('La metrica F1 para el primer modelo:',metrics.f1_score(y_pred_vot_May, y_test))
print('La metrica Accuracy para el primer modelo:',metrics.accuracy_score(y_pred_vot_May, y_test))
```

_Analisis:_  
El hecho de realizar el ensamble permitio mejorar un poco el accuracy de prediccion, aunque sea muy poco.  
Esto me demuestra que el hecho de ensamblar muchos modelos, permite compensar la falla o error de algunos en particular.

### Punto 8 - Votación Ponderada
